```python
import pyzx as zx
from helpers import graphical_partial_trace, sample_bitstrings
#Entanglement detection example
nq = 20 # number of qubits
d = 30   # depth of the circuit
g = zx.generate.cliffordT(nq, d)
g.apply_state('0' * nq)
zx.full_reduce(g)
g.normalize()
rho = graphical_partial_trace(g, [])
zx.full_reduce(rho)
rho.normalize()
zx.simplify.id_simp(rho)
zx.draw(rho, True)
```

```python
for v in rho.inputs():
    next_v = next(iter(rho.neighbors(v)))
    if rho.qubit(next_v) != rho.qubit(v):
        print(f"Entanglement detected at vertex {v} on qubit {rho.qubit(v)} with qubit {rho.qubit(next_v)}")
        if rho.qubit(next(iter(rho.neighbors(v)))) != rho.qubit(v):
            print(f"Entanglement detected at vertex {v} on qubit {rho.qubit(v)} with qubit {rho.qubit(next_v)}")
```
